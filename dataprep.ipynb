List out data directories here

```python
datafiles = [
    'AEX',
    'ATX',
    'BEL_20',
    'BIST_100',
    'Bovespa',
    'BSE_Sensex',
    'CAC40',
    'CSE',
    'DAX_PERFORMANCE-INDEX',
    "Dow_30",
    "Euro_Stoxx_50",
    "EURONEXT_100",
    "HANG_SENG_INDEX",
    "IBEX_35",
    "IDX_Composite",
    "KOSPI_Composite_Index",
    "MOEX_Russia_Index",
    "Nasdaq",
    "Nifty_50",
    "IPC_MEXICO",
    "MERVAL",
    "Nikkei_225",
    "Russell_2000",
    "SP_500",
    "SP_NZX_50_INDEX_GROSS",
    "SP_TSX_Composite_index",
    "SMI",
    "SSE",
    "TA_35",
    "Taiwan_Weighted"
]
```

Load the dataframes into a dictionary

```python
dataframes = {
    datafile: pd.read_csv('data/' + datafile + '/data.csv')
    for datafile in datafiles
}
```

- Add data directory as name
- Drop NaN rows
- Show summary of contents

```python
for name, frame in dataframes.items():
    frame['index'] = name
    frame = frame.dropna()
    print('{}: {} rows'.format(name, len(frame)))
    print(frame.head())
    print("/n")
    print("-------------------------------------")
```
